```python
url = 'https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv'
gapminder_indicators = pd.read_csv(url)
```

```python
gapminder_indicators.head()
```

```python
df_main = pd.DataFrame()

for year in range(2000,2020):
    with open('NBA_totals_years-Copy1.pickle','rb') as f:
        years_dict = pickle.load(f)
    df = years_dict[year]
    df.index = df['Player']
    df = df.drop(df[(df.Age == 'Age')].index)
    df = df.dropna(axis = 1, how = 'any')
    df_pos = df['Pos']
    df = df[['3P','3PA','PTS','2P','2PA']]
    df = df.astype(int)
    df['year'] = int(year)
    df['Player'] = df.index
    df['Pos'] = df_pos #.astype('category').cat.codes
    size_mapping = {
    'SG': 1,
    'SF': 1,
    'PG': 1,
    'C': 2,
    'PF': 2,
    'SF-SG': 1,
    'SF-PF': 1,
    'PF-SF': 2,
    'C-PF': 2,
    'SG-PG': 1,
    'PG-SG': 1,
    'SG-SF': 1,
    'PF-C': 2,
    'PG-SF': 1,
    'SG-PF': 1}
    df['Pos'] = df['Pos'].map(size_mapping)
    df_main = df_main.append(df)
df = df_main
df = df.dropna(axis = 1, how = 'any')
# dfext = pd.DataFrame({'{}*{}'.format(a, b): df[a].astype(float) * df[b].astype(float) for a, b in combinations(df.columns, 2)})
df_new = df
```
